```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
filename=r"D:\dataanalyst\Data Analyst Course\S34 - Data Visualization\L5\bar_chart_data.csv"
df_used_cars=pd.read_csv(filename)
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],height=df_used_cars["Cars Listings"], color=['r', 'g', 'b', 'w', 'y', 'm', 'c'])
plt.xticks(rotation=45)
plt.show() 
```

```python
#here in the x axis the labels are overlapped so we can ressolve it by increasing the size
#of the plot.the default size is 6.4*4.8
#with the figure function and figsize parameter we can control the size of each plot.
# we can also rotate the labels by using xticks
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],height=df_used_cars["Cars Listings"], color="midnightblue")
plt.xticks(rotation=45,fontsize=13)
plt.yticks(fontsize=13)
plt.title("Cars Listings By Brands",fontsize=16,fontweight="bold")
plt.ylabel("Number of Listings",fontsize=13)
plt.show() 
plt.savefig("USedcarsbar.png")
```

# Homework data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
filename=r"D:\dataanalyst\Data Analyst Course\S34 - Data Visualization\L5\bar_chart_homework_data.csv"
df_homework_data=pd.read_csv(filename)
df_homework_data
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=df_homework_data["Cities"],height=df_homework_data["Frequency"],color="#2B4570")
plt.title("Cities and frequencies",fontsize=18,fontweight="bold")
plt.xticks(fontweight="bold")
plt.yticks(fontweight="bold")
plt.xlabel("Cities",fontsize=15,fontweight="bold")
plt.ylabel("Frequencies",fontsize=15,fontweight="bold")
plt.show()
```
