```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
filename=r"D:\dataanalyst\Data Analyst Course\S34 - Data Visualization\L12\stacked_area_chart_data.csv"
df_fuel_engine_types=pd.read_csv(filename)
df_fuel_engine_types
```

```python
colors=["#011638","#7E2987","#EF2026"]
fuel_labels=["Gas","Petrol","Diesel"]  # Changed variable name from 'labels' to 'fuel_labels'
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Gas"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Diesel"],
              colors=colors,
              edgecolor='none'
             )

plt.legend(labels=fuel_labels,loc="upper left")  # Using the renamed variable
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.ylabel("Numbers of cars",fontsize=13)
plt.title("Popularity of engine fuel types",fontsize=13,weight="bold")
sns.despine()
plt.show()
```

# Homework data

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
filename=r"D:\dataanalyst\Data Analyst Course\S34 - Data Visualization\L12\area_chart_homework_data.csv"
df_area=pd.read_csv(filename)
df_area
```

```python

colors1=["#011638","#7E2987","#EF2026"]
labels1=["International","Domestic","Services"]
sns.set_style('white')
plt.figure(figsize=(12,6))
plt.stackplot(df_area["Date"],df_area["International"],df_area["Domestic"],df_area["Services"],colors=colors1,edgecolor="none")
plt.xticks(df_area["Date"],rotation=45,fontsize=13,weight="bold")
plt.yticks(fontsize=13,weight="bold")
plt.legend(loc="upper left",labels=labels1)
sns.despine() #to remove the black lines around the graph
plt.show()
```
