```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
filename=r"D:\dataanalyst\Data Analyst Course\S34 - Data Visualization\L8\pie_chart_data.csv"
df_fuel_engine_type=pd.read_csv(filename)
```

```python
df_fuel_engine_type
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(x=df_fuel_engine_type["Number of Cars"],
        labels=df_fuel_engine_type["Engine Fuel Type"].values,
        autopct='%.2f%%',
       textprops={'size':'x-large',
                   'fontweight':'bold',
                 'rotation':30,
                 'color':'w'})
plt.title("Cars by engine type",fontsize=18,fontweight='bold')
plt.legend()
plt.show()
plt.savefig("piechart.png")
```
